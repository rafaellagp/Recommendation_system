```python
import pandas as pd
import numpy as np
```

```python
ratings = pd.read_csv('/Users/rafaellaporto/BeCode/what-should-i-wacth/data.tsv', sep='\t')
tbasics = pd.read_csv('/Users/rafaellaporto/BeCode/what-should-i-wacth/data-5.tsv', sep='\t', low_memory=False)
```

```python
tbasics_movies = tbasics[tbasics.titleType == 'movie']

tbasics_movies = tbasics_movies[tbasics_movies.isAdult == '0']

tbasics_movies = tbasics_movies.drop(['originalTitle', 'titleType', 'isAdult', 'endYear'], axis=1)
```

```python
# merge tbasics_movies and ratings on tconst
tbasics_movies = tbasics_movies.merge(ratings, on='tconst')
tbasics_movies = tbasics_movies.set_index('tconst')
```

```python
tbasics_movies.startYear = tbasics_movies.startYear.replace('\\N', np.NaN).astype(float)

tbasics_movies.runtimeMinutes = tbasics_movies.runtimeMinutes.replace('\\N', np.NaN).astype(float)
```

```python
tbasics_movies = pd.concat([tbasics_movies.drop('genres', 1), tbasics_movies['genres'].str.get_dummies(sep=",").add_prefix('genre_').astype(bool)], 1)
```

```python
tbasics_movies
```

```python
tbasics_movies = tbasics_movies.drop(['genre_\\N'], axis=1)
```

```python
tbasics_movies = tbasics_movies.dropna(subset=['startYear', 'runtimeMinutes'])
```

```python
tbasics_movies.shape
```

```python
tbasics_movies[tbasics_movies['genre_Adult'] == True].count()
```

```python
drop_true = [True]
adult = tbasics_movies[tbasics_movies['genre_Adult'].isin(drop_true)]
tbasics_movies.drop(adult.index, axis=0,inplace=True)
# tbasics_movies.drop(delIndex, axis=0)

```

```python
tbasics_movies[tbasics_movies['genre_Adult'] == True]
```

```python
# convert startYear, runtimeMinutes to int
tbasics_movies.startYear = tbasics_movies.startYear.astype(np.int16)
tbasics_movies.runtimeMinutes = tbasics_movies.runtimeMinutes.astype(np.int32)
tbasics_movies.numVotes = tbasics_movies.numVotes.astype(np.int32)
tbasics_movies.averageRating = tbasics_movies.averageRating.astype(np.float16)
```
